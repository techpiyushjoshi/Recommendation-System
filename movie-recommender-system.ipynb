```python
# Lets create a helper function
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
# the issue is that the dictionary is a string
        
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
    
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# to apply stemming on data we need to install nltk
# natural language processing library
%pip install nltk 
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Now we will create a helper function
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
# Now we will have to do text vectorization
# We now have tags, user will give a movie, and we have to suggest 5 movies similar to the exisitng one
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# We now have to calculate similarity between the two similarities
# We have to calculate similiarity between the two texts
# Similar words calculattion is there, not a good way, that's why we need vectorization

# id|name|tags 
# 5000 data
# now we need to convert data into tags

# in a 2D Space of X and Y, vector is nothing but a coordinate
# like (1,4),(2,3)
# Every text will be converted to vector. Every movie will be a vector
# If someone likes a specific movie then we calculate closest vectors 

# this process is called Text Vectorization
# Methods to do these
# 1. Bag of Words (SImplest method)
# 2. TfIdf are methods to do this
# 3. Word to Vect

# Every movie has a tag, all tags are combined, it becomes close to 5 lakh words\
# large_text= tag1 + tag2 +..
# in that the most used words
# Frequency of everyword is calculated. Top 5000 word frequency is calcualted and are extracted

# Word1 |Word2 | Word3.... | WOrd 5000
# If action is the most frequently used word
# In movie if the word action comes the most
# Movie 1 --> Action (5/5000) + fantasy (4/1000) + delta (0/5000)
# Movie 2 --> Same is done for all movies
# We will get the data frame of (5000,5000)

# If the data has just two frequent words it will have 5000*2 movies

# Same maths is there in two dimensions
# Its our discretion to choose the size, more dimensionality more issues

# While doing Vectorization we don't use Stop Words
# Stop Words are those words which are used to form sentence but they don't contribute in their meaning
# like are, and, or. We remove them first
# This can be done manually
# We will use scikit-learn
# We will use the count vectorizer (class) and do the work easily


```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # by default the object that is returned is scipy array
```

```python
# We have a total of 4806 movies, we also have the same number of vectors
# in every vector there are 5000 vectors
# every movie is a vector in that 5000 dimensional space
# the more the distance the lesser the similarity
# we won't calculate Euclidean Distance but we will calculate Cosine Distance
# 0 degree (more similar) | 90 less | 180 opposite
# distance is inversely proportional to similarity
# more the dimension when we talk about Euclidean Fails, Cosine is more successful there
# (Part of Curse of Dimensionality)
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
# similairity is oopposite of distance
```

```python
similarity = cosine_similarity(vectors)
```

```python
# the shape is logical as each movie is compared against 4806 movies
# every movie is calculate against 4806
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
# sorting is done on index position basis
# we need to preserve the index postition this can be done by enumerate function
```

```python
# the diagonal of this movie will always be one
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(i,new_df.iloc[i[0]].title)
        # we will fetch poster from i
```

```python
# to recommend we first need to find the index of movie in similarity index
# then we will sort the index in decreasing order and will fetch the top 5
```

```python
recommend('Batman Begins')
```
